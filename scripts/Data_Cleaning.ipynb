```python
import pandas as pd
import os

# Load the raw data
raw_file_path = "data/raw/owid-covid-data.csv"
df = pd.read_csv(raw_file_path)

# Show first 5 rows to understand the data
df.head()
```

```python
# Keep only important columns
columns_to_keep = ['location', 'date', 'total_cases', 'new_cases', 'total_deaths']
df = df[columns_to_keep]

# Drop rows where 'location' is missing
df = df.dropna(subset=['location'])

# Fill missing values with 0
df[['total_cases', 'new_cases', 'total_deaths']] = df[['total_cases', 'new_cases', 'total_deaths']].fillna(0)

# Create new column: active_cases
df['active_cases'] = df['total_cases'] - df['total_deaths']

# Avoid negative active cases
df['active_cases'] = df['active_cases'].apply(lambda x: max(x, 0))

# Save cleaned data
os.makedirs('data/processed', exist_ok=True)
processed_file_path = "data/processed/cleaned_covid_data.csv"
df.to_csv(processed_file_path, index=False)

print(f"✅ Data cleaned and saved to {processed_file_path}")
```
